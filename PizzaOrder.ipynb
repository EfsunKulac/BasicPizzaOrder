```python
import csv
import datetime

# Menu.txt dosyası oluşturma
menu_text = """* Lütfen Bir Pizza Tabanı Seçiniz:
1: Klasik
2: Margarita
3: TürkPizza
4: Sade Pizza
* ve seçeceğiniz sos:
11: Zeytin
12: Mantarlar
13: Keçi Peyniri
14: Et
15: Soğan
16: Mısır
* Teşekkür ederiz!"""
with open("Menu.txt", "w") as file:
    file.write(menu_text)

# Üst sınıf pizza oluşturma
class Pizza:
    def __init__(self):
        self._description = "Pizza"
        self._cost = 0
        
    def get_description(self):
        return self._description
    
    def get_cost(self):
        return self._cost

# Alt sınıflar oluşturma
class KlasikPizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Klasik Pizza"
        self._cost = 100.0
        
class MargaritaPizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Margarita Pizza"
        self._cost = 120.0
        
class TurkPizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Türk Pizza"
        self._cost = 140.0
        
class SadePizza(Pizza):
    def __init__(self):
        super().__init__()
        self._description = "Sade Pizza"
        self._cost = 160.0

# Üst sınıf Decorator oluşturma
class Decorator(Pizza):
    def __init__(self, pizza):
        super().__init__()
        self._pizza = pizza
        
    def get_description(self):
        return self._description + " " + self._pizza.get_description()
    
    def get_cost(self):
        return self._pizza.get_cost() + self._cost

# Sos sınıfları oluşturma
class Zeytin(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Zeytinli"
        self._cost = 10.0
        
class Mantarlar(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Mantarlı"
        self._cost = 20.0
        
class KeciPeyniri(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Keçi Peynirli"
        self._cost = 30.0
        
class Et(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Etli"
        self._cost = 40.0
        
class Sogan(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Soğanlı"
        self._cost = 50.0
        
class Misir(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self._description = "Mısırlı"
        self._cost = 60.0

def main():
    # Menu
    print("MENÜ:")
    with open('Menu.txt', 'r') as menu_file:
        menu = menu_file.read()
        print(menu)

    # Pizza seçimi
    pizza_choice = input("Lütfen bir pizza seçin (1-4): ")
    while pizza_choice not in ['1', '2', '3', '4']:
        pizza_choice = input("Geçersiz seçim! Lütfen bir pizza seçin (1-4): ")

    # Sos seçimi
    sos_choice = input("Lütfen bir sos seçin (11-16): ")
    while sos_choice not in ['11', '12', '13', '14', '15', '16']:
        sos_choice = input("Geçersiz seçim! Lütfen bir sos seçin (11-16): ")

    # Toplam fiyatı hesapla
    pizza = None
    sos = None
    if pizza_choice == '1':
        pizza = KlasikPizza()
    elif pizza_choice == '2':
        pizza = MargaritaPizza()
    elif pizza_choice == '3':
        pizza = TurkPizza()
    elif pizza_choice == '4':
        pizza = SadePizza()

    if sos_choice == '11':
        sos = Zeytin(pizza)
    elif sos_choice == '12':
        sos = Mantarlar(pizza)
    elif sos_choice == '13':
        sos = KeciPeyniri(pizza)
    elif sos_choice == '14':
        sos = Et(pizza)
    elif sos_choice == '15':
        sos = Sogan(pizza)
    elif sos_choice == '16':
        sos = Misir(pizza)

    total_price = sos.get_cost()
    print("Toplam ücret:", total_price)
    # Kullanıcı bilgilerini topla
    name = input("Lütfen adınızı girin: ")
    tc = input("Lütfen TC Kimlik Numaranızı girin: ")
    card_num = input("Lütfen kredi kartı numaranızı girin: ")
    card_password = input("Lütfen kredi kartı şifrenizi girin: ")
    from datetime import datetime
    now = datetime.now()
    current_time = now.strftime("%Y-%m-%d %H:%M:%S")

    # Sipariş verilerini dosyaya kaydet
    with open('Orders_Database.csv', 'a') as orders_file:
        now = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
        order = f"{name}, {tc}, {card_num}, {card_password}, {sos.get_description()}, {total_price}, {current_time}\n"
        orders_file.write(order)

    print(sos.get_description(), "siparişiniz alındı! Teşekkür ederiz.")

main()
```
